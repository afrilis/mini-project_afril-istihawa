## c. Extract XML
Berikut adalah data set yang digunakan:
- population_data.xml

```python
# Memebaca file XML 
with open("datasource/population_data.xml", "r") as f:
    xml_data = f.read()

# Parse XML menggunakan Beautiful Soup
soup = BeautifulSoup(xml_data, "xml")

# Extrak data dari XML dan disimpan pada record
data = []

for record in soup.find_all("record"):
    record_data = {}

    for field in record.find_all("field"):
        name = field.get("name")
        key = field.get("key")
        value = field.text

        if name == "Country or Area":
            record_data["Country Name"] = value
            record_data["Country Code"] = key
        elif name == "Item":
            record_data["Indicator Name"] = value
            record_data["Indicator Code"] = key
        else:
            record_data[name] = value

    data.append(record_data)

# Ubah list ke DataFrame
pop3_df = pd.DataFrame(data)

pop3_df
```

## e. Extract API
Berikut adalah endpoint yang digunakan:

API: http://api.worldbank.org/v2/countries/ (URL API dari World Bank untuk data populasi semua negara)

Misal:
http://api.worldbank.org/v2/countries/all/indicators/SP.POP.TOTL/?format=json&per_page=1000

```python
url = "http://api.worldbank.org/v2/countries/all/indicators/SP.POP.TOTL/?format=json&per_page=1000"

# Mengambil data dari URL
response = requests.get(url)
data = response.json()
total_pages = data[0]['pages']

# Membuat list untuk menyimpan semua data
all_data = []

# Mengakses data dari API untuk semua data
for page in range(1, total_pages + 1):
    page_url = f"{url}&page={page}"
    response = requests.get(page_url)
    data = response.json()
    all_data.extend(data[1])

# Membuat DataFrame dari data yang diambil
api_df = pd.json_normalize(all_data)

# Simpan DataFrame ke file CSV 
api_df.to_csv("datacleaning/api_url.csv", index=False)

# Tampilkan Data
api_df
```

## Penyesuaian Data

###  Struktur Data 

Berikut adalah penyesuaian data yang dilakukan:

- Pembersihan Data, 
- Penyesuaian Kolom, 
- Ubah Tipe Data Kolom, 
- Filtering berdasarkan tanggal.

Dilakukan secara keseluruhan dengan menggunakan fungsi dan penerapan pada kondisi tertentu

```python
def data_preprocessing(data_files):
    for filename, df in data_files.items():
        # Menghapus tanda koma (",") dari isi dataset
        df = df.replace({',': ''}, regex=True)
        # Menghapus tanda spasi dari isi dataset
        df = df.apply(lambda x: x.map(lambda y: y.replace(' ', '') if isinstance(y, str) else y))

        if 'countryname' in df.columns:
            df['countryname'] = df['countryname'].str.split(';').str[0]

        if 'closingdate' in df.columns:
            df['closingdate'] = pd.to_datetime(df['closingdate'])

        if 'boardapprovaldate' in df.columns:
            # Mengubah tipe data kolom boardapprovaldate menjadi datetime
            df['boardapprovaldate'] = pd.to_datetime(df['boardapprovaldate'])
            # Mengubah NaN menjadi 0 setelah konversi ke integer
            df['boardapprovaldate'] = df['boardapprovaldate'].dt.year.fillna(0).astype(int)

        columns_to_clean = ['lendprojectcost', 'ibrdcommamt', 'idacommamt', 'totalamt', 'grantamt']
        for col in columns_to_clean:
            if col in df.columns:
                # Menghilangkan koma dari kolom yang ditentukan
                df[col] = df[col].replace({',': ''}, regex=True)
                # Mengubah NaN menjadi 0
                df[col] = df[col].fillna(0).astype('int64')

        # Filtering Data Berdasarkan Tanggal
        if 'date' in df.columns:
            df = df[(df['date'] >= '2018') & (df['date'] <= '2023')]
            # Mengubah nama kolom
            df.rename(columns={'countryiso3code': 'Country Code', 'date': 'Year', 'value': 'Population', 'country.value': 'Country Name'}, inplace=True)
            # Mereset index dan menghapus nama kolom
            df = df.pivot(index=['Country Name', 'Country Code'], columns='Year', values='Population').reset_index()
            df.columns_name = None
            # Menghapus tanda "," dari isi dataset
            df = df.replace({',': ''}, regex=True)
            # Menghapus tanda " " dari isi dataset
            df = df.apply(lambda x: x.map(lambda y: y.strip() if isinstance(y, str) else y))

        # Menimpa DataFrame yang sudah dimodifikasi ke dalam data_files
        data_files[filename] = df

        # Menyimpan DataFrame yang dimodifikasi ke dalam folder datacleaning, menimpa versi sebelumnya
        df.to_csv(f"datacleaning/{filename}_clean.csv", index=False)

        # Menampilkan informasi tentang DataFrame setelah preprocessing
        print(f"\nDataFrame setelah preprocessing untuk file: {filename}")
        display(df.info())
        display(df)

    return data_files

# Gunakan fungsi data_preprocessing untuk melakukan penyesuaian pada DataFrame
data_files_processed = data_preprocessing(data_files)
```

### Nilai yang hilang

- Menghapus kolom dengan nilai yang hilang lebih dari 50%
- Mengatasi nilai yang hilang dengan bfiil dan ffill

```python
def handle_missing_values(data_files, threshold=0.5):
    # Loop melalui setiap DataFrame
    for filename, df in data_files.items():
        # Menghitung jumlah baris dalam DataFrame
        total_rows = len(df)
        # Menghitung jumlah missing value dalam setiap kolom
        missing_values = df.isnull().sum()
        # Menghitung proporsi missing value dalam setiap kolom
        missing_values_ratio = missing_values / total_rows
        # Filter kolom yang memiliki proporsi missing value lebih dari threshold
        columns_to_drop = missing_values_ratio[missing_values_ratio > threshold].index
        # Menghapus kolom-kolom tersebut dari DataFrame
        df = df.drop(columns=columns_to_drop)
        
        # Mengatasi missing value dengan backward fill (bfill) dan forward fill (ffill)
        df = df.ffill().bfill()  # Menggunakan ffill() dan bfill() untuk mengatasi missing value
        
        # Menimpa DataFrame yang sudah dimodifikasi ke dalam data_files
        data_files[filename] = df

        # Menyimpan DataFrame yang dimodifikasi ke dalam folder datacleaning, menimpa versi sebelumnya
        df.to_csv(f"datacleaning/{filename}_clean.csv", index=False)
        
        # Menampilkan informasi tentang DataFrame setelah penanganan missing value
        print(f"\nDataFrame setelah penanganan missing value untuk file: {filename}")
        display(df.info())
        display(df)
        
        # Mengecek apakah DataFrame mengalami perubahan setelah penanganan missing value
        if df.equals(data_files[filename]):
            print(f"Tidak ada perubahan yang terjadi setelah penanganan missing value untuk file: {filename}")
        
    return data_files

# Gunakan fungsi handle_missing_values untuk menangani missing value pada DataFrame yang sudah di-preprocess
data_files_processed = handle_missing_values(data_files_processed)
```

### Dummy Variabel

Penyelesaian termasuk dalam Pemeriksaan Data

```python
def handle_dummy_variables(data_files):
    # Pemanggilan dummy variabel
    for filename, df in data_files.items():
        for column in df.select_dtypes(include=['object']).columns:
            dummy_variables = pd.get_dummies(df[column], prefix=column)
            df = pd.concat([df, dummy_variables], axis=1)

    # Menampilkan data sebelum dan sesudah penambahan variabel dummy
    for filename, df in data_files.items():
        print(f"\nData sebelum penambahan variabel dummy untuk {filename}:")
        display(df)  # Menampilkan data sebelum penambahan variabel dummy

        print(f"\nData setelah penambahan variabel dummy untuk {filename}:")
        display(df)  # Menampilkan data setelah penambahan variabel dummy

    return data_files

# Gunakan handle_dummy_variables untuk melakukan penyesuaian pada DataFrame
data_files_processed = handle_dummy_variables(data_files)
```
